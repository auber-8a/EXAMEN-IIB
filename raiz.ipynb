```python
import numpy as np
import matplotlib.pyplot as plt

# Función
def f(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

# Método de bisección
def biseccion(a, b, tol=1e-12):
    if f(a) * f(b) >= 0:
        print("El intervalo [a, b] no contiene una raíz")
        return None
    c = a
    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c


# Graficamos la funcion para ver los intervalos
x = np.linspace(-5, 5, 1000)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = x^5 - 6x^4 + 2x^3 + 20x^2 - 27x + 10', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.ylim(-100, 100)
plt.xlim(-5, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función')
plt.grid(True)
plt.legend()
plt.show()

# Intervalos identificados al ojimetro
intervalos = [(-4, 0), (0, 2), (2, 6), (6, 1000)]
 
for intervalo in intervalos:
    a, b = intervalo
    raiz = biseccion(a, b)
    if raiz is not None:
        print(f"La raíz aproximada en el intervalo [{a}, {b}] es {raiz:.12f}")
```

Como podemos ver, solo existen 3 raices. Que es donde pasa la gráfica por el eje x.
